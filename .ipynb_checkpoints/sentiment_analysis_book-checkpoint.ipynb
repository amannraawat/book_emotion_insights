# Book Sentiment Analysis Project

## Load the Book

```python
with open("2-states_the-story-of-my-marriage-chetan-bhagat.txt", "r", encoding="utf8") as file:
    book = file.read()
```

```python
book
```

## How many acts are there in the story?

```python
#with string method
book.count("Act")
```

```python
#with regular expression
import re
pattern = re.compile("Act [0-9]")
re.findall(pattern, book)
```

## Sentences with which 'love' and  'sad' are used


```python
pattern = re.compile("[^.]*[^a-zA-Z]+sad[^a-zA-Z]+[^.]*.")
re.findall(pattern, book)
```

```python
pattern = re.compile("sad")
re.findall(pattern, book)
```

```python
pattern = re.compile("[^.]* love [^.]*.")
re.findall(pattern, book)
```
